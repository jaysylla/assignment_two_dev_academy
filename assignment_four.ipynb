```python
"""
Lost in Lagos
"""

# ------------------------------
# Room Map
# ------------------------------
rooms = {
    "Airport": {
        "description": "You just landed at Murtala Muhammed Airport. Welcome! The air is hot and it's chaotic.",
        "directions": {"east": "Oshodi"}
    },
    "Oshodi": {
        "description": "The busy streets are filled with danfo buses. It's loud here.",
        "directions": {"west": "Airport", "north": "Yaba", "east": "Ikeja"}
    },
    "Ikeja": {
        "description": "Ikeja hasn't changed in years. You spot the famous Computer Village nearby.",
        "directions": {"west": "Oshodi", "north": "Lekki"}
    },
    "Yaba": {
        "description": "Tech hubs and markets line the streets of Yaba. You spot Paga's office nearby.",
        "directions": {"south": "Oshodi", "east": "Victoria Island"}
    },
    "Lekki": {
        "description": "Traffic is building up, especially towards the link bridge.",
        "directions": {"south": "Ikeja", "east": "Banana Island"}
    },
    "Victoria Island": {
        "description": "Skyscrapers and Ozumba Mbadiwe. You're close to your destination!",
        "directions": {"west": "Yaba", "south": "Banana Island"}
    },
    "Banana Island": {
        "description": "You've made it to Banana Island! Welcome home",
        "directions": {"north": "Lekki", "west": "Victoria Island"},
        "is_goal": True
    }
}

# ------------------------------
# Game Functions
# ------------------------------
def play_game():
    current_room = "Airport"
    steps = 0
    
    # üìú Game Instructions
    print("\nüìú Welcome to Lost in Lagos!\n")
    print("üìú How to Play:")
    print("1Ô∏è‚É£ Your goal is to find Banana Island.")
    print("2Ô∏è‚É£ You will start at Murtala Muhammed Airport and can move through different parts of Lagos.")
    print("3Ô∏è‚É£ Each location has a description and available directions (e.g., north, south, east, west).")
    print("4Ô∏è‚É£ Type the direction you want to move and press Enter.")
    print("5Ô∏è‚É£ Type 'quit' anytime to exit the game.")
    print("6Ô∏è‚É£ The game tracks how many moves it takes you to find the goal.")
    print("üí° Enjoy exploring Lagos!\n")
    
    print("üõ¨ You have arrived in Lagos, Nigeria. Let us begin!\n")

    while True:
        room = rooms[current_room]
        print(f"\nüìç Location: {current_room}")
        print(room["description"])

        directions = room.get("directions", {})
        if not directions:
            print("üöß No exits from here. You're stuck!")
            break

        print("Available directions:", ', '.join(directions.keys()))
        move = input("Which direction do you want to go? ").lower()

        if move == "quit":
            print("\nüëã You chose to quit the game. Goodbye!")
            break

        if move in directions:
            current_room = directions[move]
            steps += 1
        else:
            print("‚ùå Can't go that way. Try another route.")

        if rooms.get(current_room, {}).get("is_goal"):
            print(f"\nüéâ {rooms[current_room]['description']}")
            print(f"üéØ You reached Banana Island in {steps} moves!")
            break

def ask_replay():
    while True:
        replay = input("\nüîÅ Would you like to play again? (yes/no): ").lower()
        if replay in ["yes", "y"]:
            return True
        elif replay in ["no", "n"]:
            return False
        else:
            print("Please enter 'yes' or 'no'.")

def main():
    while True:
        play_game()
        if not ask_replay():
            print("üëã Thanks for exploring Lagos. Enjoy your stay!")
            break

# Start the game
if __name__ == "__main__":
    main()
```
