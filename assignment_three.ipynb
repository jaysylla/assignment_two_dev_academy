```python
print("Welcome to Megan's Madlibs!")
print("These are the rules:")
print("1. You will be asked to provide a series of words.")
print("2. The words will be used to fill in the blanks of a story.")
print("3. At the end, you will see the completed story.")
print("4. Have fun and be creative!")
print("Let's get started!")
print("If you want to quit at any time, just type 'quit'.")

while True:
    noun1 = input("Enter a noun: ")
    if noun1.lower() == "quit":
        print("Thanks for playing!")
        break

    noun2 = input("Enter another noun: ")
    if noun2.lower() == "quit":
        print("Thanks for playing!")
        break

    noun3 = input("Enter a third noun: ")
    if noun3.lower() == "quit":
        print("Thanks for playing!")
        break

    verb1 = input("Enter a verb: ")
    if verb1.lower() == "quit":
        print("Thanks for playing!")
        break

    verb2 = input("Enter another verb: ")
    if verb2.lower() == "quit":
        print("Thanks for playing!")
        break

    verb3 = input("Enter a third verb: ")
    if verb3.lower() == "quit":
        print("Thanks for playing!")
        break

    adjective1 = input("Enter an adjective: ")
    if adjective1.lower() == "quit":
        print("Thanks for playing!")
        break

    adjective2 = input("Enter another adjective: ")
    if adjective2.lower() == "quit":
        print("Thanks for playing!")
        break

    adjective3 = input("Enter a third adjective: ")
    if adjective3.lower() == "quit":
        print("Thanks for playing!")
        break

    # Story output
    story = f"""
My weekend was eventful. It all started when I decided to go to the {noun1} and I saw a {noun1} that was {verb1} through the forest.
Along the way, I was feeling a little bit {adjective1} and to make myself feel {adjective2} I decided to go to {noun2}.
After I {verb2} I felt {adjective3}, it got late, so I decided to go to {noun3}, I was {verb3}.
It was tiring, and eventful, but I'm looking forward to a new week!
"""
    print("\nHereâ€™s your story!")
    print(story)
    print("Thanks for playing Megan's Madlibs!")
    break  # done with game!
```
